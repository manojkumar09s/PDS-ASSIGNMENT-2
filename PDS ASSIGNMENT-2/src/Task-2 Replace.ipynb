```python
# Loading New Dataset
import pandas as pd
new_data = pd.read_csv('/content/drive/MyDrive/PDS ASSIGNMENT-2/task1.csv')
```

```python
# Remove units ( kmpl, CC, bhp) from  columns

new_data['Mileage'] = new_data['Mileage'].str.replace(' kmpl', '')
new_data['Mileage'] = new_data['Mileage'].str.replace(' km/kg', '')
new_data['Engine'] = new_data['Engine'].str.replace(' CC', '')
new_data['Power'] = new_data['Power'].str.replace(' bhp', '')


```
