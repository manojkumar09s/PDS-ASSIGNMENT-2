```python
# Load the dataset

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/PDS ASSIGNMENT-2/raw_data.csv')
```

```python

# Counting the  null values
null_counter = df.isnull().sum()
print(null_counter)
```

From above Output we can say that the null values are present in the columns of: "Mileage" "Engine" "Power" "Seats" "New_Price"

So, we need to clean the data from above respective columns.

```python

modified_columns = ['Mileage', 'Engine', 'Power', 'Seats']

for column in modified_columns:
    if df[column].dtype == np.float64 or df[column].dtype == np.int64:
        df[column].fillna(df[column].mean(), inplace=True)
    elif df[column].dtype == object:
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Dropping the columns which have high range of missing values

df.drop(columns=['New_Price'], inplace=True)
```

```python
miss_values = df.isnull().sum()
print("let's see the missing values")
print(miss_values)
```

```python
# Saved the modified data to a new CSV file
df.to_csv('/content/drive/MyDrive/PDS ASSIGNMENT-2/finalOutput.csv')
```
