```python
# Loading New Dataset
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/PDS ASSIGNMENT-2/finalOutput.csv')
```

```python
#select
selectquery_df = df[['Name', 'Year', 'Price']]
print(selectquery_df)
```

```python
#filter
filterquery_df = df[(df['Year'] > 2014) & (df['Fuel_Type'] == 'Petrol')]
print(filterquery_df)
```

```python
# renaming the columns
df = df.rename(columns={'Transmission': 'Gear_Type'})
print(df)
```

```python
#mutate
df['Car_Stock'] = 2020 - df['Year']
print(df)
```

```python
#arrange
arrangequery = df.sort_values(by='Price', ascending=False)
print(arrangequery)
```

```python
#Summaeize
summaryquery = df.groupby('Fuel_Type').agg(Average_Price=('Price', 'mean'), Average_Age=('Car_Stock', 'mean')).reset_index()
print(summaryquery)
```
